```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
data_cancer = load_breast_cancer()
```

```python
dir(data_cancer)
```

```python
df = pd.DataFrame(data_cancer.data,columns=data_cancer.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
df['target'] = data_cancer.target
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler_data = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
x_pca = PCA(n_components=2)
data_pca = x_pca.fit_transform(scaler_data)
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression()
model.fit(data_pca,data_cancer.target)
```

```python
model.score(data_pca,data_cancer.target)
```

```python
model.fit(scaler_data,data_cancer.target)
```

```python
model.score(scaler_data,data_cancer.target)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0], data_pca[:,1], c=data_cancer.target)
plt.xlabel('First_PC')
plt.ylabel('Second_PC')
```
